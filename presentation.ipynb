```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
import numpy as np

fig, ax = plt.subplots()
x = np.linspace(0, 10, 100)

def update(freq):
    ax.clear()
    ax.plot(x, np.sin(freq * x))
    fig.canvas.draw_idle()

freq_slider = widgets.FloatSlider(value=1, min=0.1, max=5, step=0.1, description='Frequenz')
widgets.interactive(update, freq=freq_slider)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

def update_plots(freq, amp):
    ax1.clear()
    ax2.clear()
    ax1.plot(x, amp * np.sin(freq * x))
    ax2.plot(x, amp * np.cos(freq * x))
    ax1.set_title('Sinus')
    ax2.set_title('Cosinus')
    fig.canvas.draw_idle()

freq_slider = widgets.FloatSlider(value=1, min=0.1, max=5, step=0.1, description='Frequenz')
amp_slider = widgets.FloatSlider(value=1, min=0.1, max=2, step=0.1, description='Amplitude')

widgets.interactive(update_plots, freq=freq_slider, amp=amp_slider)
```
